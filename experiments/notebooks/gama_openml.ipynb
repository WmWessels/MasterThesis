```python
import openml
import traceback
from gama import GamaClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, accuracy_score
import logging
```

```python
openml_df = openml.datasets.list_datasets(output_format="dataframe")
classification_datasets = openml_df[openml_df['NumberOfClasses'] >= 2].index
```

```python
logging.basicConfig(filename='results/axuiliary_errors.log', level=logging.DEBUG, 
                    format='%(asctime)s %(levelname)s %(name)s %(message)s')
logger_self =logging.getLogger(__name__)
```

```python
for ind in classification_datasets[5:]:
    dataset = openml.datasets.get_dataset(ind)
    data, _, _, _ = dataset.get_data(dataset_format="dataframe")
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)    
        automl = GamaClassifier(max_total_time=100, output_directory = f"results/logs_{ind}", store="logs")
        automl.fit(X_train, y_train)
    except Exception as e:
        logger_self.error(e)
```

```python
label_predictions = automl.predict(X_test)
probability_predictions = automl.predict_proba(X_test)
```

```python
print('accuracy:', accuracy_score(y_test, label_predictions))
print('log loss:', log_loss(y_test, probability_predictions))
# the `score` function outputs the score on the metric optimized towards (by default, `log_loss`)
print('log_loss', automl.score(X_test, y_test))
```
