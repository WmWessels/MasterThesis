```python
import openml
import traceback
from gama import GamaClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, accuracy_score
import logging
```

```python
openml_df = openml.datasets.list_datasets(output_format="dataframe")
classification_datasets = openml_df[openml_df['NumberOfClasses'] >= 2].index
```

```python
for ind in classification_datasets:
    dataset = openml.datasets.get_dataset(ind)
    data, _, _, _ = dataset.get_data(dataset_format="dataframe")
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)    
        automl = GamaClassifier(max_total_time=100, output_directory = f"results/logs_{ind}", store="logs")
        automl.fit(X_train, y_train)
    except Exception:
        pass
```
